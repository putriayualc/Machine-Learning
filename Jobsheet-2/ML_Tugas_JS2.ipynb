# Deskripsi Tugas  
Pada tugas pratikum ini Anda akan menggunakan data "Wisconsin Breast Cancer". Data tersebut terdiri dari 569 data yang digunakan untuk mendiagnonis jenis kanker Malignant (M) dan Benign (B). Tugas Anda adalah,
1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

2. Lakukan proses encoding pada kolom "diagnosis".

3. Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

4. Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

## Load data

```python
import pandas as pd

data = '/content/wbc.csv' # path dataset
df = pd.read_csv(data) # load dataset

df.head()
```

Kode tersebut memuat dataset CSV (wbc.csv) ke dalam sebuah DataFrame (df) menggunakan pandas dan menampilkan 5 baris pertama dari data tersebut.  
pandas adalah library yang digunakan untuk manipulasi data dalam bentuk tabel (DataFrame) di Python. DataFrame adalah struktur data 2 dimensi seperti tabel, di mana setiap kolom bisa memiliki tipe data yang berbeda (misalnya angka, string, dll).   
metode .head() pada DataFrame df digunakan untuk menampilkan 5 baris pertama dari dataset. Ini berguna untuk mendapatkan gambaran awal tentang data yang dimuat, termasuk kolom apa saja yang ada dan bagaimana bentuk datanya.

## Tugas 1
Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
df = df.drop(columns=['id', 'Unnamed: 32']) #digunakan untuk menghapus kolom id dan Unnamed
df.head() #menampilkan 5 baris atas
```

Kode tersebut digunakan untuk membersihkan dataset dengan menghapus dua kolom yang tidak diperlukan, yaitu id dan Unnamed: 32, dari DataFrame df. Setelah penghapusan kolom, metode df.head() digunakan untuk menampilkan lima baris pertama dari DataFrame yang sudah dibersihkan, memungkinkan pengguna untuk memeriksa perubahan dan memastikan bahwa kolom yang dihapus tidak lagi ada dalam data.

## Tugas 2
Lakukan proses encoding pada kolom 'diagnosis'

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

le = LabelEncoder()                                     # membuat objek dari LabelEncoder
df['diagnosis'] = le.fit_transform(df['diagnosis'])     # proses encoding

df.head()
```

Kode tersebut melakukan proses encoding pada kolom diagnosis dalam DataFrame df. Pertama, LabelEncoder diimpor dari sklearn.preprocessing dan objek LabelEncoder dibuat. Kemudian, kolom diagnosis diubah dari format kategori menjadi format numerik menggunakan metode fit_transform dari LabelEncoder.

## Tugas 3
Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
std = StandardScaler()
scale = df.columns.difference(['diagnosis'])
df[scale] = std.fit_transform(df[scale])

df.head()
```

Kode tersebut melakukan standarisasi pada kolom-kolom numerik dalam DataFrame df, kecuali kolom diagnosis. Pertama, objek StandardScaler dibuat untuk menangani proses standarisasi. Kemudian, daftar nama kolom yang akan distandarisasi diperoleh dengan mengecualikan kolom diagnosis. Selanjutnya, kolom-kolom yang terpilih tersebut distandarisasi menggunakan metode fit_transform dari objek StandardScaler. Hasil standarisasi kemudian menggantikan data asli dalam kolom-kolom tersebut, memastikan bahwa semua kolom numerik berada dalam skala yang konsisten.

## Tugas 4
Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
# Split data
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])

# Cek masing-masing ukuran data
print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}')
print(f'Jumlah label data train:\n{df_train.diagnosis.value_counts()}')
print(f'Jumlah label data test:\n{df_test.diagnosis.value_counts()}')
```

Kode tersebut digunakan untuk membagi DataFrame df menjadi data latih (df_train) dan data uji (df_test). Dengan menggunakan fungsi train_test_split, data dibagi dengan rasio 80:20, di mana 20% dari data digunakan sebagai data uji dan sisanya 80% sebagai data latih. Parameter stratify=df['diagnosis'] memastikan bahwa distribusi kelas dalam kolom diagnosis tetap proporsional di kedua subset data. Setelah pembagian, kode memeriksa dan menampilkan jumlah distribusi label di data asli, data latih, dan data uji untuk memastikan bahwa pembagian data telah dilakukan dengan benar dan distribusi kelas tetap konsisten.
